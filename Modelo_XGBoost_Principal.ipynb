## Processamento de dados

```python
df = pd.read_csv("Ibovespa_01.01.2015__04.07.2025.csv")
```

```python
df.head()
```

## Limpeza e correção dos dados

```python
# Função para padronizar os dados
def corrige_volumetria(valor:str):
    if isinstance(valor, str):
        valor = valor.replace(".", "").replace(",", ".")
        if "K" in valor:
            return float(valor.replace("K", "")) * 1_000
        elif "M" in valor:
            return float(valor.replace("M", "")) * 1_000_000
        elif "B" in valor:
            return float(valor.replace("B", "")) * 1_000_000_000
        elif "%" in valor:
            return float(valor.replace("%", ""))
    return float(valor)
```

```python
# Renomeação das colunas para simplificar
colunas = {
    "Último": "Fechamento",
    "Abertura": "Abertura",
    "Máxima": "Maxima",
    "Mínima": "Minima",
    "Vol.": "Volume",
    "Var%": "Variacao"
}
df = df.rename(columns=colunas)
```

```python
# Aplicação das correções dos dados
df["Data"] = pd.to_datetime(df["Data"], format="%d.%m.%Y")
df["Volume"] = df["Volume"].apply(corrige_volumetria)
df["Variacao"] = df["Variacao"].apply(corrige_volumetria)
```

```python
df.head()
```

```python
# Heatmap para veificar multicolinearidade
plt.figure(figsize=(12, 8))
sns.heatmap(df.drop(columns=["Data"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação das Variáveis")
plt.show()
```

## Feature Engineering

```python
# Aplicação de Feature Engineering
df = df.sort_values("Data").reset_index(drop=True)
df["Target"] = (df["Fechamento"].shift(-1) > df["Fechamento"]).astype(int)
df["Retorno_Diario"] = df["Fechamento"].pct_change()
df["MM_3"] = df["Fechamento"].rolling(window=3).mean()
df["MM_5"] = df["Fechamento"].rolling(window=5).mean()
df["MM_10"] = df["Fechamento"].rolling(window=10).mean()
df["MM_15"] = df["Fechamento"].rolling(window=15).mean()
df["Volatilidade_5"] = df["Fechamento"].rolling(window=5).std()
df["Spread_Day"] = df["Maxima"] - df["Minima"]
df["Fechamento_Abertura"] = df["Fechamento"] - df["Abertura"]

df = df.dropna().reset_index(drop=True)
```

```python
df.tail()
```

```python
# Plotagem da distribuição dos targets 
sns.countplot(x=df["Target"])
plt.title("Distribuição do Target (Subiu = 1 / Caiu = 0)")
plt.show()
```

```python
# Lista de features/colunas que não serão usadas no modelo
feature_drop = [
    "Data", "Abertura"
    ]
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.drop(columns=feature_drop).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação das Variáveis")
plt.show()
```

## Modelagem e normalização

```python
# Lista de features/colunas que serão usadas no modelo
features = [
    "Fechamento", "Volume", "Retorno_Diario", "MM_3", "MM_5", "MM_10", "MM_15",
    "Volatilidade_5", "Spread_Day", "Fechamento_Abertura"
]

x = df[features]
y = df["Target"]

# Normalização dos dados
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Separação do time series para 30 dias de treino
x_train = x_scaled[:-30]
x_test = x_scaled[-30:]
y_train = y[:-30]
y_test = y[-30:]
```

```python
sns.countplot(x=df["Target"][-30:])
plt.title("Distribuição do Target (Subiu = 1 / Caiu = 0)")
plt.show()
print(df["Target"][-30:].count())
```

```python
# Treinamento do modelo XGBoost Classifier
param_grid = {
    'n_estimators': [0, 50, 100, 150, 200],
    'max_depth': [0, 3, 5, 10],
    'learning_rate': [0.05, 0.1],
    'alpha': [0, 0.5, 1],   # L1 regularization
    'lambda': [1, 2, 5]     # L2 regularization
}

# Modelo base
xgb_model = XGBClassifier(
    random_state=42
)

# Configuração do GridSearch
grid = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=3,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)

# Treino com cross-validation
grid.fit(x_train, y_train)

print("🔧 Melhores parâmetros encontrados:")
print(grid.best_params_)

# Melhor modelo já treinado
best_model = grid.best_estimator_

y_pred = best_model.predict(x_test)
#{'alpha': 0, 'lambda': 1, 'learning_rate': 0.05, 'max_depth': 10, 'n_estimators': 150}
```

```python
print(classification_report(y_test, y_pred))
print("Acurácia nos últimos 30 dias:", f"{accuracy_score(y_test, y_pred) * 100:.2f}%")
```

```python
ConfusionMatrixDisplay.from_estimator(best_model, x_test, y_test, cmap="Blues")
plt.title("Matriz de Confusão")
plt.grid(False)
plt.show()
```

```python
# Features mais importantes do modelo
importancia = pd.Series(best_model.feature_importances_, index=features).sort_values()
importancia.plot(kind="barh", figsize=(10, 6), title="Importância das Features (XGBoost)")
plt.xlabel("Importância")
plt.show()
```

## Previsão da bolsa para o próximo dia

```python
# Ao invés de criar coluna por coluna no DataFrame abaixo, crio um dicionário para criar igual no X quando recebe o DataFrame com as features
ultimo = df.iloc[-1]
x_novo_dict = {col: ultimo[col] for col in features}
x_novo = pd.DataFrame([x_novo_dict])[features] # DataFrame criado com o dicionário e as features de X
x_novo_scaled = scaler.transform(x_novo)
```

```python
# Previsão e Probabilidade do modelo
prev = best_model.predict(x_novo_scaled)[0]
proba = best_model.predict_proba(x_novo_scaled)[0]
print(f"Dia seguinte do pregão da bolsa 07/07/2025: {'Vai subir!' if prev == 1 else 'Vai cair!'}")
print(f"Probabilidades - Cair: {proba[0]:.2%}, Subir: {proba[1]:.2%}")
```
